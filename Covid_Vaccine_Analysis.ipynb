```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('covid_vaccine_statewise.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.fillna(0)
```

```python
first_dose = df.groupby('State')['First Dose Administered'].max()
print(first_dose)
```

```python
second_dose = df.groupby('State')['Second Dose Administered'].max()
print(second_dose)
```

```python
total_male = df['Male (Doses Administered)'].sum()
print("Total Male Vaccinated:", total_male)
```

```python
total_female = df['Female (Doses Administered)'].sum()
print("Total Female Vaccinated:", total_female)
```

```python
import matplotlib.pyplot as plt

first_dose.sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 States - First Dose")
plt.show()
```
